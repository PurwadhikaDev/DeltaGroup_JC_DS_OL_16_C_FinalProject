## Metric Evaluation:
- Mean Absolute Error (MAE)
Mengukur rata-rata perbedaan absolut antara nilai aktual dan nilai prediksi. MAE menunjukkan seberapa besar kesalahan prediksi dalam satuan data yang sebenarnya. Semakin kecil nilai MAE, semakin akurat modelnya.

- Mean Absolute Percentage Error (MAPE)
Mengukur rata-rata kesalahan prediksi dalam bentuk persentase terhadap nilai aktual. Berguna untuk memahami akurasi prediksi secara relatif. Semakin kecil nilai MAPE, semakin baik model dalam memprediksi data.

- Root Mean Squared Error (RMSE)
Mengukur akar rata-rata dari kuadrat kesalahan prediksi. RMSE memberikan penalti yang lebih besar untuk kesalahan besar, sehingga lebih sensitif terhadap outlier dibandingkan MAE. Nilai RMSE yang lebih kecil menunjukkan model yang lebih akurat.

### Outlying Data

*Outlier* perlu ditindak jika data akan digunakan dalam menentukan pemodelan. Adanya _outlier_ bisa memengaruhi hasil pemodelan menjadi kurang optimal karena titik-titik _outlier_ tersebut mau tidak mau tetap dipertimbangkan pada saat model dibuat. Terdapat satu kemungkinan jenis _outlier_ pada kasus ini, yakni secara global saja, yakni yang bisa dibuktikan dengan statistika. _Outlier_ kontekstual yang perlu tampilan visual hubungan variabel dependen dan independen yang berkaitan tidak bisa diejawantahkan pada kasus ini karena bisa saja variabel dependen dipengaruhi oleh lebih dari satu variabel independen, dan _outlier_ kolektif yang mempertimbangkan semua data sangat terkait seperti data _time-series_.

```python
df_numeric_columns = df_clean.select_dtypes(include = np.number).columns.tolist()

print('Outliers based on IQR and quantiles')
for column in df_numeric_columns:
    Q1 = df_clean[column].quantile(0.25)
    Q3 = df_clean[column].quantile(0.75)
    IQR = Q3 - Q1
    lowerBorder = Q1 - 1.5 * IQR
    upperBorder = Q3 + 1.5 * IQR
    print(f'Outliers in {column}, size: {df_clean[(df_clean[column] < lowerBorder) | (df_clean[column] > upperBorder)].shape[0]}, percentage: {round(df_clean[(df_clean[column] < lowerBorder) | (df_clean[column] > upperBorder)].shape[0] / df_clean.shape[0] * 100, 1)} %')
    display(df_clean[(df_clean[column] < lowerBorder) | (df_clean[column] > upperBorder)])


df_numeric = df_clean.select_dtypes(include = [np.number])
z_scores = np.abs(scist.zscore(df_numeric))
df_no_outliers = df_clean[(z_scores < 3).all(axis = 1)]
print('Outliers based on z-score')
print(f'Data frame without outliers, size: {df_no_outliers.shape[0]}, percentage: {round(df_no_outliers.shape[0] / df_clean.shape[0] * 100, 1)} %')
display(df_no_outliers)
print(f'Outliers found at least in one column, size: {df_clean.shape[0] - df_no_outliers.shape[0]}, percentage: {round((df_clean.shape[0] - df_no_outliers.shape[0]) / df_clean.shape[0] * 100, 1)} %')
```

Dapat disimak bahwa ternyata hanya 5,0 % dari total keseluruhan baris yang memuat nilai _outlier_ setidaknya pada satu kolom numerik. Dengan demikian, menghapus baris dengan nilai _outlier_ ini tidak menjadi masalah berarti, meskipun mungkin saja proporsi hasil analisis pasca-pembersihan menjadi berubah dengan hilangnya data _outlier_ seperti pada harga jual dan luas unit apartemen. Serta, untuk pemilihan data mana saja yang perlu _to be omiited_, daripada dengan perhitungan konvensional berupa batas atas dan batas bawah yang mempertimbangkan jarak antarkuartil, lebih baik memilih pertimbangan menggunakan _z-score_ yang menggunakan besaran rerata dan standar deviasi, dan jika besaran tersebut di atas, titik data tersebut bisa dianggap sebagai _outlier_ seperti yang tertera pada [link ini](https://www.geeksforgeeks.org/z-score-in-statistics/).

```python
# Define algoritma yang digunakan
lr = LinearRegression()
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor(random_state=1)
rf = RandomForestRegressor(random_state=1)
xgb = XGBRegressor(random_state=1)

# Pemodelan dilakukan dalam skala logaritmik, namun kemudian di-inverse kembali untuk interpretasi
log_lr = TransformedTargetRegressor(lr, func=np.log, inverse_func=np.exp)
log_knn = TransformedTargetRegressor(knn, func=np.log, inverse_func=np.exp)
log_dt = TransformedTargetRegressor(dt, func=np.log, inverse_func=np.exp)
log_rf = TransformedTargetRegressor(rf, func=np.log, inverse_func=np.exp)
log_xgb = TransformedTargetRegressor(xgb, func=np.log, inverse_func=np.exp)


# Kandidat algoritma
models = [log_lr, log_knn, log_dt, log_rf, log_xgb]

score_rmse = []
nilai_mean_rmse = []
nilai_std_rmse = []

score_mae = []
nilai_mean_mae = []
nilai_std_mae = []

score_mape = []
nilai_mean_mape = []
nilai_std_mape = []

# Mencari algoritma terbaik berdasarkan tiap metrics
for i in models:

    crossval = KFold(n_splits=5, shuffle=True, random_state=1)

    estimator = Pipeline([
        ('preprocessing', preprocessor),

        ('model', i)
    ])

    # RMSE
    model_cv_rmse = cross_val_score(
        estimator,
        X_train,
        y_train,
        cv=crossval,
        scoring='neg_root_mean_squared_error',
        error_score='raise'
        )

    print(model_cv_rmse, i)

    score_rmse.append(model_cv_rmse)
    nilai_mean_rmse.append(model_cv_rmse.mean())
    nilai_std_rmse.append(model_cv_rmse.std())

    # MAE
    model_cv_mae = cross_val_score(
        estimator,
        X_train,
        y_train,
        cv=crossval,
        scoring='neg_mean_absolute_error',
        error_score='raise'
        )

    print(model_cv_mae, i)

    score_mae.append(model_cv_mae)
    nilai_mean_mae.append(model_cv_mae.mean())
    nilai_std_mae.append(model_cv_mae.std())

    # MAPE
    model_cv_mape = cross_val_score(
        estimator,
        X_train,
        y_train,
        cv=crossval,
        scoring='neg_mean_absolute_percentage_error',
        error_score='raise'
        )

    print(model_cv_mape, i)

    score_mape.append(model_cv_mape)
    nilai_mean_mape.append(model_cv_mape.mean())
    nilai_std_mape.append(model_cv_mape.std())
```

Perbedaan ini bisa terjadi karena performa model pada data validasi (cross-validation) tidak selalu mencerminkan performa pada data uji (test set). Ada beberapa faktor yang bisa menyebabkan XGBoost tampil lebih baik pada test set meskipun terlihat kurang baik saat cross-validation. Berikut adalah beberapa kemungkinan penyebabnya:

 1. Variabilitas dalam Cross-Validation

- Cross-validation membagi data menjadi beberapa subset (folds) untuk evaluasi performa model.
- Ada kemungkinan bahwa dalam fold tertentu, XGBoost bekerja kurang optimal karena data yang dilatih memiliki pola yang tidak sepenuhnya sesuai.
- Namun, ketika diuji pada test set (yang memiliki distribusi berbeda), XGBoost bisa lebih mampu mengenali pola dengan baik, sehingga hasilnya lebih baik dibandingkan RandomForest.

2. Random Forest Lebih Mudah Overfitting

- Random Forest terdiri dari banyak decision tree yang cenderung lebih cepat
overfitting jika jumlah pohonnya terlalu besar.
- XGBoost menggunakan boosting, yang lebih terkontrol dalam belajar dari kesalahan, sehingga cenderung memiliki generalisasi lebih baik pada data yang tidak terlihat sebelumnya.
-Jika Random Forest memiliki performa lebih baik dalam cross-validation tetapi lebih buruk di test set, itu bisa menjadi indikasi overfitting.


3. Skala dan Variabilitas Data Uji
- Cross-validation menggunakan berbagai subset dari data latih, sedangkan pengujian model akhir dilakukan pada data yang benar-benar baru (test set).
- Jika distribusi data test set berbeda dari distribusi pada training set, model yang lebih fleksibel (seperti XGBoost) mungkin lebih mampu beradaptasi dengan perbedaan tersebut dibandingkan Random Forest.

**Hyperparameter Tuning**

Karena model XGBoost memiliki performa yang sedikit lebih baik dari RandomForest, maka selanjutnya kita akan menggunakan model XGBoost ini sebagai model akhir. Pada tahap ini, kita akan melakukan hyperparameter tuning pada model XGBoost dengan harapan dapat meningkatkan performa model.

Untuk panduan tuning model XGBoost, dapat merujuk ke [Tuning XGBoost](https://xgboost.readthedocs.io/en/latest/parameter.html#general-parameters).

```python
# Kedalaman pohon
max_depth = list(np.arange(1, 11))

# Learning rate
learning_rate = list(np.arange(1, 100)/100)

# Jumlah pohon
n_estimators = list(np.arange(100, 201))

# Jumlah baris tiap pohon (% dari total baris train set)
subsample = list(np.arange(2, 10)/10)

# Gamma (min_impurity_decrease)
gamma = list(np.arange(1, 11)) # Semakin besar nilainya, semakin konservatif/simpel modelnya

# Jumlah feature yang digunakan untuk tiap pohon (% dari total kolom train set)
colsample_bytree = list(np.arange(1, 10)/10)

# Alpha (regularization)
reg_alpha = list(np.logspace(-3, 1, 10)) # Semakin besar nilainya, semakin konservatif/simpel modelnya


# Hyperparam space XGboost
hyperparam_space_xgb = {
    'model__max_depth': max_depth,
    'model__learning_rate': learning_rate,
    'model__n_estimators': n_estimators,
    'model__subsample': subsample,
    'model__gamma': gamma,
    'model__colsample_bytree': colsample_bytree,
    'model__reg_alpha': reg_alpha
}
```

```python
# Benchmark model dengan hyperparameter tuning
xgb = XGBRegressor(random_state=1, verbosity=0)

# Membuat algorithm chains
estimator_xgb = Pipeline([
        ('preprocessing', preprocessor),
        ('model', xgb)
        ])

crossval = KFold(n_splits=5, shuffle=True, random_state=1)

# Hyperparameter tuning
random_xgb = RandomizedSearchCV(
    estimator_xgb,
    param_distributions = hyperparam_space_xgb,
    n_iter = 50,
    cv = crossval,
    scoring = ['neg_root_mean_squared_error', 'neg_mean_absolute_error', 'neg_mean_absolute_percentage_error'],
    n_jobs = -1,
    refit = 'neg_root_mean_squared_error', # Hanya bisa memilih salah stau metric untuk optimisasi
    random_state = 1
)
```

```python
# Fitting data training untuk mencari parameter terbaik
random_xgb.fit(X_train, y_train)
```

```python
# Melihat hasil tuning dalam bentuk dataframe. Diurutkan berdasarkan RMSE, MAE & MAPE
pd.DataFrame(random_xgb.cv_results_).sort_values(by=['rank_test_neg_root_mean_squared_error', 'rank_test_neg_mean_absolute_error', 'rank_test_neg_mean_absolute_percentage_error']).head()
```

```python
# Cek skor dan params terbaik
print('XGBoost')
print('Best_score:', random_xgb.best_score_)
print('Best_params:', random_xgb.best_params_)
```

**Predict to Test Set with the Tuned Model**

Melakukan prediksi pada test set dengan menggunakan model XGBoost dan hyperparameter terpilih.

```python
# Model XGBoost
model = {'XGB': XGBRegressor(random_state=1)}

# Define model terhadap estimator terbaik
xgb_tuning = random_xgb.best_estimator_

# Fitting model
xgb_tuning.fit(X_train, y_train)

# Predict test set
y_pred_xgb_tuning = xgb_tuning.predict(X_test)

# Simpan nilai metrics RMSE, MAE & MAPE setelah tuning
rmse_xgb_tuning = np.sqrt(mean_squared_error(y_test, y_pred_xgb_tuning))
mae_xgb_tuning = mean_absolute_error(y_test, y_pred_xgb_tuning)
mape_xgb_tuning = mean_absolute_percentage_error(y_test, y_pred_xgb_tuning)

score_after_tuning = pd.DataFrame({'RMSE': rmse_xgb_tuning, 'MAE': mae_xgb_tuning, 'MAPE': mape_xgb_tuning}, index=model.keys())
score_after_tuning
```

Setelah dilakukan hyperparameter tuning, RMSE XGBoost membaik (dari 11.090 menjadi 10.782), tetapi MAE dan MAPE justru sedikit memburuk (MAE naik dari 6.285 menjadi 6.395, dan MAPE naik dari 0.0928 menjadi 0.0947). Ini bisa terjadi karena beberapa alasan berikut:

1.Tuning Hyperparameter Bisa Mengorbankan Satu Metric untuk Memperbaiki yang Lain.
- Hyperparameter tuning sering kali mencoba meningkatkan performa dalam satu aspek, tetapi bisa mengorbankan aspek lain.
- Dalam hal ini, tuning berhasil menurunkan RMSE, yang berarti prediksi model lebih baik dalam menangani outlier atau variabilitas yang besar.
- Namun, MAE dan MAPE justru naik sedikit, yang menunjukkan bahwa kesalahan rata-rata dalam prediksi angka yang lebih kecil mungkin meningkat.

RMSE lebih sensitif terhadap outlier, sehingga tuning mungkin lebih fokus mengurangi kesalahan besar, tetapi ini bisa menyebabkan peningkatan MAE dan MAPE pada prediksi lainnya.

2.Trade-off antara Bias dan Variance
- Sebelum tuning, model mungkin masih agak overfitting, sehingga cenderung memiliki MAE yang lebih rendah tetapi RMSE yang lebih tinggi.
- Setelah tuning, model mungkin menjadi lebih generalisasi, yang mengurangi RMSE tetapi mengorbankan sedikit akurasi dalam rata-rata kesalahan absolut (MAE dan MAPE).

3.Dataset Test Bisa Berperan dalam Perubahan Ini
- Jika dataset test memiliki distribusi nilai harga yang tidak merata, tuning hyperparameter bisa lebih baik dalam mengurangi error besar (RMSE) tetapi tidak terlalu membantu dalam kesalahan absolut rata-rata (MAE, MAPE).
